###**i. The difference in np.array, np.asarray and np.asanyarray.**

Here's a detailed explanation of the differences between `np.array`, `np.asarray`, and `np.asanyarray`:

###**np.array**

np.array is the most commonly used function to create a NumPy array. It takes an object as input and returns a new NumPy array. If the input is already a NumPy array, a copy of the array is returned.

Here are some key features of `np.array`:

    * Creates a new array, even if the input is already a NumPy array.
    * Copies the data from the input object.
    * Can take a wide range of input types, including lists, tuples, and other NumPy arrays.
    * Returns a new array with the specified dtype (if provided).

   **EXAMPLE**


```python
import numpy as np

myList = [1, 2, 3, 4, 5]
arr = np.array(myList)

print(arr)
```

###**np.asarray**

`np.asarray` is similar to `np.array`, but it returns a view of the input array if it's already a NumPy array. This means that if the input is already a NumPy array, `np.asarray` returns the same array object, without copying the data.

Here are some key features of `np.asarray`:

    * Returns a view of the input array if it's already a NumPy array.
    * Does not copy the data if the input is already a NumPy array.
    * Can take a wide range of input types, including lists, tuples, and other NumPy arrays.
    * Returns an array with the same dtype as the input array.

**Example:**


```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])
arrView = np.asarray(arr)
print(arrView)
print(arr is arrView)
```

###**np.asanyarray**

`np.asanyarray` is similar to `np.asarray`, but it returns an array with a specific dtype if the input is not already a NumPy array. This is useful when you want to ensure that the output array has a specific dtype, even if the input is not a NumPy array.

Here are some key features of `np.asanyarray`:

    
    * Returns a view of the input array if it's already a NumPy array.
    * Does not copy the data if the input is already a NumPy array.
    * Can take a wide range of input types, including lists, tuples, and other NumPy arrays.
    * Returns an array with the specified dtype (if provided).

**Example:**

```python
import numpy as np

myList = [1, 2, 3, 4, 5]
arr = np.asanyarray(myList, dtype=np.float64)
print(arr)
print(arr.dtype)
```

###**ii. The difference between Deep copy and shallow copy**

##**Shallow Copy**

A `shallow copy` of an object is a new object that references the same memory locations as the original object. This means that both objects share the same underlying data. When you modify the copied object, you're actually modifying the original object as well, because they're pointing to the same memory locations.

In Python, you can create a shallow copy of an object using the copy module:

```python
import copy

originalList = [[1, 2], [3, 4]]
shallowCopy = copy.copy(originalList)
print(shallowCopy)

shallowCopy[0][0] = 10
print(originalList)
```

##**Deep Copy**

A `deep copy` of an object is a new object that recursively creates new copies of all the objects it references. This means that the copied object has its own independent copy of the data, and modifying the copied object does not affect the original object.

In Python, you can create a deep copy of an object using the copy module:

```python
import copy

originalList = [[1, 2], [3, 4]]
deepCopy = copy.deepcopy(originalList)

print(deepCopy)
deepCopy[0][0] = 10
print(originalList)
```

## **14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and Perform the following tasks using Matplotlib:**

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dateRange = pd.date_range(start='2023-01-01', periods=100, freq='D')

np.random.seed(0)
temperature = np.random.uniform(low=15.0, high=35.0, size=(100,))
humidity = np.random.uniform(low=30.0, high=90.0, size=(100,))
df = pd.DataFrame({
    'Date': dateRange,
    'Temperature': temperature,
    'Humidity': humidity
})

fig, ax1 = plt.subplots(figsize=(10, 5))

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='tab:red')
ax1.plot(df['Date'], df['Temperature'], color='tab:red', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:red')

ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='tab:blue')
ax2.plot(df['Date'], df['Humidity'], color='tab:blue', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:blue')

plt.title('Temperature and Humidity Over Time')

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.tight_layout()
plt.show()
```

## **15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:**

a) Plot a histogram of the data with 30 bins.

b) Overlay a line plot representing the normal distribution's probability density function(PDF).

c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.

d) Set the title of the plot as 'Histogram with PDF Overlay'.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(data, bins=30, density=True, alpha=0.5, label='Histogram')

x = np.linspace(data.min(), data.max(), 100)
plt.plot(x, norm.pdf(x, loc=0, scale=1), 'r-', label='Normal Distribution PDF')

plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

plt.title('Histogram with PDF Overlay')

plt.legend()

plt.show()
```

## **17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.rand(100) * 20 - 10
y = np.random.rand(100) * 20 - 10

def get_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Quadrant I'
    elif x < 0 and y > 0:
        return 'Quadrant II'
    elif x < 0 and y < 0:
        return 'Quadrant III'
    elif x > 0 and y < 0:
        return 'Quadrant IV'
    else:
        return 'Origin'

data = pd.DataFrame({'x': x, 'y': y})
data['Quadrant'] = data.apply(lambda row: get_quadrant(row['x'], row['y']), axis=1)

plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(data=data, x='x', y='y', hue='Quadrant', palette='Set1', style='Quadrant', markers=["o", "s", "D", "^"])
plt.legend(title='Quadrants')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.show()
```

## **18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.**

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

p = figure(title='Sine Wave Function', x_axis_label='X', y_axis_label='sin(X)', width=800, height=400)

p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

p.line(x, y, line_width=2, color='blue', legend_label='sin(X)')

p.legend.location = "top_left"

show(p)
```

## **19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.**

```python
from bokeh.models import HoverTool
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

output_notebook()

categories = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H', 'I', 'J']
values = np.random.randint(1, 20, size=len(categories))

p = figure(x_range=categories, title='Random Categorical Bar Chart',
           x_axis_label='Category', y_axis_label='Value', width=800, height=400)

mapper = linear_cmap(field_name='top', palette=Viridis256, low=min(values), high=max(values))

bars = p.vbar(x=categories, top=values, width=0.8, fill_color=mapper, line_color='black')

hover = HoverTool(tooltips=[('Category', '@x'), ('Value', '@top')])
p.add_tools(hover)

show(p)
```
